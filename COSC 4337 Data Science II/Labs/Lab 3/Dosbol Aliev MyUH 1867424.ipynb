Here is the real meat of the lab. You need ot understand backpropagation in order to complete the "train_network" function. This function will contain our calculation of errors and changing of the weights and biases.

In this particular Lab we want to make a neural network with a 2-5-1 architecture with 3 layers. You can see in the feed_forward function we have these calculations. In this case we want to do a simular calculation to lab2/HW1 and then modify the value of the weights in Train_network.

To start by calculating errors we want to calculate the error of the output layer by taking the error between the predicted and true value and multiple that by the derivative of the sigmoid function. The derivative can be calcualted in the sigmoid_d function. We want to do the same thing with the hidden layer but in this case we consider the dot product of the output layer and output gradient, as well as the sigmoid derivative of the output of the hidden layer.

With these two in hand we have everything we need to modify the weights and biases. In this case the biases can be changed by summing the gradient variables, and the weights can be modified by using a dot product on the input of that layer * the gradient of the relavent layer.
Don't forget to multiple everything by the learning rate!

```python
class Advanced_Perceptron():
    def __init__(self, num_input, num_hidden, num_output, alpha): # 2-5-1 architecture
        
        self.num_input = num_input
        self.num_hidden = num_hidden
        self.num_output = num_output
        self.alpha = alpha #learning rate
        
        self.weights_h = np.random.random(size=(num_input, num_hidden))
        self.bias_h = random.random()

        self.weights_o = np.random.random(size=(num_hidden, num_output))
        self.bias_o = random.random()
        
    def sigmoid(self, x_train):
        return 1 / (1 + np.exp(-x_train))

    def sigmoid_d(self, y_output):#Sigmoid derivative
        return  y_output* (1 - y_output)

    def total_error(self, y_true, y_pred):
        return .5 * (y_true - y_pred)**2

    def error_d(self, y_true, y_pred): # error function, may be useful in calculating 
        return (y_true - y_pred)

    def feed_forward(self, inputs):
        hidden_layer = np.dot(inputs,self.weights_h)
        hidden_layer += self.bias_h
        y_hidden = self.sigmoid(hidden_layer)
        
        output_layer= np.dot(y_hidden,self.weights_o)
        output_layer +=  self.bias_o
        y_output = self.sigmoid(output_layer)
        
        return y_output, y_hidden
    
    

    def train_network(self, inputs, outputs):
        #print(x_train)
        y_output, y_hidden = self.feed_forward(inputs)
        
        E = self.total_error(y_train, y_output).sum()#.mean()
        
        #back propagation
        #Calculate the amount by which we want to change the weights and biases
        E1=self.error_d(outputs,y_output)
        gradient_o =  E1* self.sigmoid_d(y_output)
        
        F1=gradient_o.dot(self.weights_o.T)
        gradient_h  = F1 * self.sigmoid_d(y_hidden )
        #Next apply the changes to the weights and the biases
        
  
        self.bias_o += np.sum(gradient_o,axis=0,keepdims=True) * self.alpha
        self.bias_h += np.sum(gradient_h,axis=0,keepdims=True)* self.alpha
        self.weights_o += y_hidden.T.dot(gradient_o) * self.alpha
        self.weights_h += inputs.T.dot(gradient_h) * self.alpha
        
        return E
```
